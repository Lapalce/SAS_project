```python
col_listings = listings.columns.tolist()
col_dic = dictionary['Field'].tolist()
col_listings.sort()
col_dic.sort()
```

```python
i = 0
extra_name = []
for index, name in enumerate(col_listings):
    if i < len(col_dic):
        if name == col_dic[i]:
            print(f"{index}: {name} -- {i}:{col_dic[i]}")
            i += 1
        else :
            extra_name.append(name)
            print(f"{index}: {name}")
    else:
        print(f"{index}: {name}")
```
